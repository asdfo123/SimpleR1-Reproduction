```python
# 导入必要的库
from vllm import LLM, SamplingParams

# 定义模型名称，此处使用 Qwen-2.5-Instruct
model_name = "Qwen2.5-1.5B"

# 加载 VLLM 的 LLM 引擎
# 如果下载模型速度较慢，可以尝试设置本地缓存路径
# tensor_parallel_size = 2  # 使用 2 张 GPU（根据实际设备数量调整）

# 加载 VLLM 的 LLM 引擎
# 通过设置 tensor_parallel_size 启用多卡
llm = LLM(model=model_name)

# 定义一个简单的输入提示
prompt = "Up to isomorphism, how many additive abelian groups G of order 16 have the property that x + x + x + x = 0 for each x in G ? choices:[\"0\",\"1\",\"2\",\"3\"]"

# 定义采样参数（可根据需求调整）
sampling_params = SamplingParams(
    temperature=0.7,  # 控制生成的随机性，0.7 是常用值
    max_tokens=512,   # 最大生成长度
    top_p=0.9,        # 对采样分布进行截断
)

# 使用 LLM 推理生成结果
outputs = llm.generate([prompt], sampling_params)

# 打印生成的结果
for output in outputs:
    print("生成的回答:")
    print(output.outputs[0].text)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# 加载模型和分词器
model_name = "/mnt/raid/lixinye/smolR1/outputs/simple_rl_zoo_qwen2.5-1.5b/checkpoint-72-output"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", trust_remote_code=True)

# 定义输入提示
prompt = "Up to isomorphism, how many additive abelian groups G of order 16 have the property that x + x + x + x = 0 for each x in G ? choices:[\"0\",\"1\",\"2\",\"3\"]"

# 对输入进行编码
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

# 生成回答
outputs = model.generate(
    **inputs,
    max_length=16000,
    temperature=0.7,
    top_p=0.9,
    pad_token_id=tokenizer.pad_token_id
)

# 解码并打印结果
response = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("生成的回答:")
print(response[len(prompt):])  # 只打印生成的部分
```
